```python
def upper(text):

    string=''

    for char in text:

        if(ord(char)>=97 and ord(char)<=122):

            string+=chr(ord(char)-32)

        else:

            string+=char

    return string

def lower(text):

    string=''

    for char in text:

        if(ord(char)>=65 and ord(char)<=90):

            string+=chr(ord(char)+32)

        else:

            string+=char

    return string



 

print(upper('delhi'))

print(lower('CARTOON'))

```

```python
def check_alpha(text):

    c=0

    for i in text:

        if((ord(i)>=65 and ord(i)<=90) or(ord(i)>=97 and ord(i)<=122)):

            c+=1

    if(len(text)==c):

        return True

    else:

        return False

print('datascience'.isalpha())

print('123'.isalpha())
```

```python
def check_digit(text):

    c=0

    for i in text:

        if((ord(i)>=48 and ord(i)<=57)):

            c+=1

    if(len(text)==c):

        return True

    else:

        return False

    

print(check_digit('datascience'))

print(check_digit('1234'))

```

```python
def upper(text):

    string=''

    for char in text:

        if(ord(char)>=97 and ord(char)<=122):

            string+=chr(ord(char)-32)

        else:

            string+=char

    return string

def lower(text):

    string=''

    for char in text:

        if(ord(char)>=65 and ord(char)<=90):

            string+=chr(ord(char)+32)

        else:

            string+=char

    return string

 

def title(text):

    output=[]

    for word in text.split(' '):

        output.append(upper(word[0])+lower(word[1:]))

    

    return ' '.join(output)




 

print(title("let us study machine learning today"))


```
